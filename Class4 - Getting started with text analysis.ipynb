```python
female = ' '.join(female_words)
f_wordcloud = WordCloud(background_color="white", max_font_size=40).generate(female)
plt.imshow(f_wordcloud)
plt.axis("off")  
```

## Sentiment Analysis

(slide 112)

```python
test = ['I am a very happy person.', 'I am a very sad person', 
        'Ive always understood happiness to be appreciation. There is no greater happiness than \
        appreciation for what one has- both physically and in the way of relationships and \
        ideologies. The unhappy seek that which they do not have and can not fully appreciate the \
        things around them. I dont expect much from life. I dont need a high paying job, a big \
        house or fancy cars. I simply wish to be able to live my life appreciating everything around me.', 
        'I love life and never sad wedge abhor']
```

The TextBlob library can detect the sentiment of a piece of text using the `sentiment` property on a TextBlob class object. The returned values include the polarity score, which is between -1 and 1, indicating the positivity and negativity of the text. 

```python
for d in test:
    d = TextBlob(d)
    print(d.sentiment)
```
